```python
# Importar las herramientas necesarias
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# crear la tabla original con datos
df = pd.DataFrame({'animal': ['perro', 'gato', 'perro', 'pez', 'gato']})
print("Tabla original:")
print(df)

# crear el codificador
# le decimos que queremos una salida normal (no comprimida)
ohe = OneHotEncoder(sparse_output=False)

# aprender las categorías únicas que hay en la columna 'animal'
ohe.fit(df[['animal']])

# usar esas categorías para transformar el texto en unos y ceros crea una matriz de 1 y 0
encoded = ohe.transform(df[['animal']])

# convertir esa matriz en una tabla legible, con nombres de columna
column_names = ohe.get_feature_names_out(['animal'])
df_encoded = pd.DataFrame(encoded, columns=column_names)
print("\n Tabla codificada (DataFrame):")
print(df_encoded)

# Paso 7: Unir la tabla original con la nueva tabla codificada
df_total = pd.concat([df, df_encoded], axis=1)
print("\n Tabla final (original + codificada):")
print(df_total)
```

## 5.4.- IMPUTACION SIN LIBRERIAS

```python
# imputacion sin librerias
#Es el proceso de rellenar datos faltantes (vacíos o NaN).
#Aquí lo hacemos manualmente, sin usar librerías automáticas como sklearn.

#Métodos comunes:
#  Reemplazar por la media (promedio).

#  Reemplazar por la mediana.

#  Reemplazar por un valor constante.

import pandas as pd
import numpy as np

# creamos una tabla con algunas edades faltantes (nan)
df = pd.DataFrame({ 'edad': [25, np.nan, 30, 22, np.nan, 28]})

print('tabla original:')
print(df)

# calcular la media (promedio) de los valores que no son nan
media = df['edad'].mean(skipna=True)

print('\nmedia calculada:', media)

#recorrer la columna y reemplazar nan por la media
nueva_columna = []
for valor in df['edad']:
    if pd.isna(valor):  # si el valor esta vacio es nan
        nueva_columna.append(media)  # agregamos la media
    else:
        nueva_columna.append(valor)  # dejamos el valor original

# actualizar la columna en el dataframe
df['edad'] = nueva_columna

print('\ntabla final con valores reemplazados:')
print(df)

```

```python
#Es un proceso ordenado que aplica varios pasos al mismo conjunto de datos.
#Ejemplo típico: primero limpiar datos, luego codificarlos, luego entrenar un modelo.

#Nosotros lo hacemos paso a paso manualmente, no con sklearn.pipeline.
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Creamos un dataset mixto: texto, numérico y nulos
df = pd.DataFrame({
    'nombre': ['Ana', 'Luis', 'Carla', 'Luis', 'Ana'],
    'edad': [23, np.nan, 31, 28, np.nan],
    'genero': ['F', 'M', 'F', 'M', 'F']
})

# Definimos una función que aplica pasos en orden
def pipeline_manual(data):
    # Reemplaza NaN en edad con la media
    data['edad'] = data['edad'].fillna(data['edad'].mean())

    # Codifica género usando LabelEncoder ('F' = 0, 'M' = 1)
    le = LabelEncoder()
    data['genero_encoded'] = le.fit_transform(data['genero'])

    return data  # Retorna el nuevo DataFrame

# Ejecuta la función con una copia del dataset original
df_final = pipeline_manual(df.copy())
print(df_final)
```

## 5.5.- PIPELINE

```python
import pandas as pd
import numpy as np

# librerias de sklearn para automatizar los pasos
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer

# datos con valores faltantes y texto
df = pd.DataFrame({
    'nombre': ['ana', 'luis', 'carla', 'pablo', 'diana'],
    'edad': [25, np.nan, 30, 22, np.nan],
    'genero': ['f', 'm', 'f', 'm', 'f']
})

print('📋 datos originales:')
print(df)

# definimos las columnas numericas y categoricas
columnas_numericas = ['edad']
columnas_categoricas = ['genero']

# pipeline para numericos (imputar con la media)
pipeline_numerico = Pipeline([
    ('imputador', SimpleImputer(strategy='mean'))
])

# pipeline para texto (codificar con ordinalencoder)
pipeline_categorico = Pipeline([
    ('codificador', OrdinalEncoder())
])

# combinamos todo con columntransformer
preprocesador = ColumnTransformer([
    ('num', pipeline_numerico, columnas_numericas),
    ('cat', pipeline_categorico, columnas_categoricas)
])

# aplicamos el pipeline al dataframe (solo columnas de interes)
datos_transformados = preprocesador.fit_transform(df)

# convertimos a tabla con nombres
df_procesado = pd.DataFrame(datos_transformados, columns=['edad', 'genero_codificado'])

# pegamos con la columna nombre
df_final = pd.concat([df[['nombre']], df_procesado], axis=1)

print('\n✅ datos despues del pipeline automatico:')
print(df_final)
```

```python
from deap import base, creator, tools
import random

# Definimos los tipos de fitness y los individuos
creator.create("Evaluacion", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.Evaluacion)

def generar_individuo():
    individuo = []
    for _ in range(10):
        gen = random.randint(0, 1)
        individuo.append(gen)
    return individuo

def evaluar(individual):
    return sum(individual),

toolbox = base.Toolbox()
toolbox.register("individual", tools.initIterate, creator.Individual, generar_individuo)
toolbox.register("poblacion", tools.initRepeat, list, toolbox.individual)
toolbox.register("evaluate", evaluar)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)

# ================== FASE DE GENERACIONES ==================
n_generaciones = 5

# GENERA LA POBLACION
poblacion = toolbox.poblacion(n=10)

for generacion in range(n_generaciones):

    print(f"\nGENERACIÓN {generacion + 1}")

    print("\n POBLACION:")
    for i, individuo in enumerate(poblacion):
        print(f"Individuo {i + 1}: {individuo}")

    # EVALUAR Y MOSTRAR
    print("\nEvaluación de la población:")
    for i, individuo in enumerate(poblacion):
        if not individuo.fitness.valid:
            individuo.fitness.values = toolbox.evaluate(individuo)  # Guardamos la aptitud
        print(f"Aptitud de Individuo {i + 1}: {individuo.fitness.values[0]}")

    # SELECCION DE LOS MEJORES INDIVIDUOS
    seleccionados = toolbox.select(poblacion, len(poblacion))

    # Cruce y Mutación
    nueva_poblacion = []
    for i in range(0, len(seleccionados), 2):
        padre1 = seleccionados[i]
        padre2 = seleccionados[i + 1]

        # EVALUAMOS ANTES DE CRUCAR
        if not padre1.fitness.valid:
            padre1.fitness.values = toolbox.evaluate(padre1)
        if not padre2.fitness.valid:
            padre2.fitness.values = toolbox.evaluate(padre2)

        # CRUCE
        if random.random() < 0.8:
            hijo1, hijo2 = toolbox.mate(padre1, padre2)
            del hijo1.fitness.values
            del hijo2.fitness.values
        else:
            hijo1 = creator.Individual(padre1)
            hijo2 = creator.Individual(padre2)
        # MUTACION

        hijo1 = toolbox.mutate(hijo1)[0]
        del hijo1.fitness.values
        hijo2 = toolbox.mutate(hijo2)[0]
        del hijo2.fitness.values

        nueva_poblacion.append(hijo1)
        nueva_poblacion.append(hijo2)

    poblacion = nueva_poblacion


final = toolbox.select(poblacion, 1)[0]
if not final.fitness.valid:
    final.fitness.values = toolbox.evaluate(final)
print("\n POSIBLE RESULTADO:")
print(f"{final} con la evaluacion {final.fitness.values[0]}")
```

```python
import random
import numpy as np

# LO QUE SE NECESITA (PAR EL DE POBLACION)
n_bits = 6
n_poblacion = 10
n_generaciones = 4

prob_cruce = 0.8
prob_mutacion = 0.1


def funcion_objetivo(x):
    return x**2 + 1

def binario_a_entero(binario):
    numero_binario = ""
    for bit in binario:
        numero_binario += str(bit)
    return int(numero_binario, 2)

def generar_individuo():
    individuo = []
    for i in range(n_bits):
        bit = random.randint(0, 1)
        individuo.append(bit)
    return individuo

def generar_poblacion():
    poblacion = []
    for i in range(n_poblacion):
        individuo = generar_individuo()
        poblacion.append(individuo)
    return poblacion

def evaluar_poblacion(poblacion):
    evaluaciones = []
    for individuo in poblacion:
        x = binario_a_entero(individuo)
        resultado = funcion_objetivo(x)
        evaluaciones.append(resultado)
    return evaluaciones

def seleccion(poblacion, fitness):
    seleccionados = []
    for vez in range(len(poblacion)):
        indice1 = random.randint(0, len(poblacion) - 1)
        indice2 = random.randint(0, len(poblacion) - 1)
        if fitness[indice1] > fitness[indice2]:
            ganador = poblacion[indice1]
        else:
            ganador = poblacion[indice2]
        seleccionados.append(ganador)
    return seleccionados

def cruzar(padre1, padre2):
    if random.random() < prob_cruce:
        punto = random.randint(1, n_bits - 1)
        hijo1 = padre1[:punto] + padre2[punto:]
        hijo2 = padre2[:punto] + padre1[punto:]
        return hijo1, hijo2
    return padre1[:], padre2[:]

def mutar(individuo):
    for i in range(len(individuo)):
        if random.random() < prob_mutacion:
            individuo[i] = 1 - individuo[i]
    return individuo




poblacion = generar_poblacion()

for generacion in range(n_generaciones):

    fitness = evaluar_poblacion(poblacion)
    print(f"\n GENERACION::: {generacion+1}")
    indice = 1
    for individuo in poblacion:
        valor_decimal = binario_a_entero(individuo)
        resultado_funcion = funcion_objetivo(valor_decimal)
        print("Individuo", indice, ":", individuo, "| x =", valor_decimal, "| f(x) =", resultado_funcion)
        indice += 1

    seleccionados = seleccion(poblacion, fitness)
    nueva_poblacion = []

    for i in range(0, n_poblacion, 2):
        padre1 = seleccionados[i]
        padre2 = seleccionados[i+1]
        hijo1, hijo2 = cruzar(padre1, padre2)
        hijo1 = mutar(hijo1)
        hijo2 = mutar(hijo2)
        nueva_poblacion.append(hijo1)
        nueva_poblacion.append(hijo2)

    poblacion = nueva_poblacion
```

# 7.- N CABALLOS

## 7.1.- ALGORITMO DE LOS N CABALLOS

```python
from deap import base, creator, tools
import random

N = 5  # NxN N Caballos

creator.create("Evaluacion", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.Evaluacion)

def generar_individuo():
    tablero = [0] * (N * N)
    posiciones = random.sample(range(N * N), N)
    for pos in posiciones:
        tablero[pos] = 1
    return tablero

def evaluar(individual):
    conflictos = 0
    tablero = individual
    for i in range(N):
        for j in range(N):
            if tablero[i * N + j] == 1:
                movimientos = [
                    (i+2, j+1), (i+2, j-1),
                    (i-2, j+1), (i-2, j-1),
                    (i+1, j+2), (i-1, j+2),
                    (i+1, j-2), (i-1, j-2)
                ]
                for x, y in movimientos:
                    if 0 <= x < N and 0 <= y < N:
                        if tablero[x * N + y] == 1:
                            conflictos += 1
    return (conflictos // 2,)

def mutar(individual):

    posiciones_caballos = [i for i, val in enumerate(individual) if val == 1]
    posiciones_vacias = [i for i, val in enumerate(individual) if val == 0]

    if posiciones_caballos and posiciones_vacias:
        caballo_a_mover = random.choice(posiciones_caballos)
        nueva_posicion = random.choice(posiciones_vacias)

        # Intercambiar
        individual[caballo_a_mover] = 0
        individual[nueva_posicion] = 1
    return (individual,)

toolbox = base.Toolbox()
toolbox.register("individual", tools.initIterate, creator.Individual, generar_individuo)
toolbox.register("poblacion", tools.initRepeat, list, toolbox.individual)
toolbox.register("evaluate", evaluar)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", mutar)
toolbox.register("select", tools.selTournament, tournsize=3)

# ================== FASE DE GENERACIONES ==================
n_generaciones = 2
poblacion = toolbox.poblacion(n=20)

for generacion in range(n_generaciones):

    print(f"\nGENERACIÓN {generacion + 1}")

    print("\n POBLACION:")
    for i, individuo in enumerate(poblacion):
        print(f"Individuo {i + 1}: {individuo}")

    print("\nEvaluación de la población:")
    for i, individuo in enumerate(poblacion):
        if not individuo.fitness.valid:
            individuo.fitness.values = toolbox.evaluate(individuo)
        print(f"Aptitud de Individuo {i + 1}: {individuo.fitness.values[0]} (conflictos)")

    seleccionados = toolbox.select(poblacion, len(poblacion))

    nueva_poblacion = []
    for i in range(0, len(seleccionados), 2):
        padre1 = seleccionados[i]
        padre2 = seleccionados[i + 1]

        if not padre1.fitness.valid:
            padre1.fitness.values = toolbox.evaluate(padre1)
        if not padre2.fitness.valid:
            padre2.fitness.values = toolbox.evaluate(padre2)

        if random.random() < 0.8:
            hijo1, hijo2 = toolbox.mate(padre1, padre2)
            del hijo1.fitness.values
            del hijo2.fitness.values
        else:
            hijo1 = creator.Individual(padre1)
            hijo2 = creator.Individual(padre2)

        hijo1 = toolbox.mutate(hijo1)[0]
        del hijo1.fitness.values
        hijo2 = toolbox.mutate(hijo2)[0]
        del hijo2.fitness.values

        nueva_poblacion.append(hijo1)
        nueva_poblacion.append(hijo2)

    poblacion = nueva_poblacion

# Mejor individuo al final
final = toolbox.select(poblacion, 1)[0]
if not final.fitness.valid:
    final.fitness.values = toolbox.evaluate(final)
print("\n POSIBLE RESULTADO:")

print(f"{final} con {final.fitness.values[0]} conflictos")
```

# 8.- PERCEPTRON PYTHON


```python
import numpy as np

# Datos de entrenamiento
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# Salida esperada
Y = np.array([0, 0, 0, 1])

# Parámetros
tasa_aprendizaje = 0.3
epocas = 200

# Inicializamos pesos (incluyendo bias)
# 3 pesos: w1 (bias), w2 (x2), w3 (x3)
pesos = np.array([1.0, 1.0, 1.0])  # puedes inicializarlos aleatorio también

# Entrenamiento
for epoca in range(epocas):
    print(f"Época {epoca+1}")
    for i in range(len(X)):
        # Entrada extendida con bias = 1
        entrada = np.array([1, X[i][0], X[i][1]])

        # Calcular la salida (función de activación escalón)
        suma = np.dot(pesos, entrada)
        y_calculado = 1 if suma >= 0 else 0

        # Error
        error = Y[i] - y_calculado

        # Actualización de pesos
        delta = tasa_aprendizaje * error * entrada
        pesos += delta

        print(f"Entrada: {entrada}, Salida esperada: {Y[i]}, Salida calculada: {y_calculado}, Pesos: {pesos}")

print("\nEntrenamiento finalizado")
print(f"Pesos finales: {pesos}")

# Prueba final
print("\nPruebas:")
for i in range(len(X)):
    entrada = np.array([1, X[i][0], X[i][1]])
    suma = np.dot(pesos, entrada)
    y_calculado = 1 if suma >= 0 else 0
    print(f"Entrada: {X[i]}, Salida: {y_calculado}")
```

```python
import numpy as np
import tensorflow as tf
import random


# clase principal del juego tres en raya
class tres_en_raya:
    # constructor, inicializa el tablero vacío
    def __init__(self):
        # crea un tablero de 3x3, inicializado en 0 (vacío)
        self.board = np.zeros((3, 3), dtype=int)  # 0 = vacío, 1 = jugador, -1 = IA

    # resetea el tablero a vacío
    def reset(self):
        self.board[:] = 0

    # muestra el tablero en consola
    def display(self):
        # diccionario que mapea los valores a símbolos visuales
        symbols = {1: 'X', -1: 'O', 0: ' '}
        # imprime cada fila del tablero
        for row in self.board:
            print('|'.join(symbols[cell] for cell in row))
            print('-'*5)

    # devuelve las posiciones disponibles para jugar
    def available_moves(self):
        return [(i, j) for i in range(3) for j in range(3) if self.board[i, j] == 0]

    # realiza un movimiento en una posición del tablero
    def make_move(self, row, col, player):
        row -= 1  # ajustamos la fila para el índice de Python (0-2)
        col -= 1  # ajustamos la columna para el índice de Python (0-2)
        if self.board[row, col] == 0:  # si la casilla está vacía
            self.board[row, col] = player  # asigna el valor del jugador en la casilla
            return True
        return False  # si la casilla ya está ocupada, no se puede mover

    # revisa si hay un ganador
    def check_winner(self):
        # revisa si hay un jugador que haya ganado
        for player in [1, -1]:
            # chequea filas, columnas y diagonales
            if any(np.all(self.board[i, :] == player) for i in range(3)) or \
               any(np.all(self.board[:, j] == player) for j in range(3)) or \
               np.all(np.diag(self.board) == player) or \
               np.all(np.diag(np.fliplr(self.board)) == player):
                return player  # devuelve el ganador (1 para jugador, -1 para IA)

        # si no hay más movimientos posibles, es empate
        if not self.available_moves():
            return 0  # empate

        return None  # si el juego sigue, no hay ganador aún


# función para crear el modelo de red neuronal
def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(9,)),  # entrada de tamaño 9 (tablero 3x3)
        tf.keras.layers.Dense(64, activation='relu'),  # capa densa con 64 neuronas
        tf.keras.layers.Dense(64, activation='relu'),  # otra capa densa con 64 neuronas
        tf.keras.layers.Dense(9, activation='softmax')  # capa de salida, con probabilidades para cada casilla
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy')  # optimizador y función de pérdida
    return model  # devuelve el modelo creado


# función que hace un movimiento aleatorio para la IA
def random_move(board):
    # obtiene todas las casillas disponibles y selecciona una al azar
    available = [(i, j) for i in range(3) for j in range(3) if board[i, j] == 0]
    return random.choice(available)  # elige una casilla al azar


# convierte el tablero 3x3 a un vector de 9 elementos (necesario para la red neuronal)
def board_to_vector(board):
    return board.flatten()  # aplana el tablero de 3x3 a un vector de 9 elementos


# convierte la salida de la red neuronal en una jugada (fila, columna)
def vector_to_move(vector):
    # devuelve la fila y columna correspondientes a la posición máxima del vector
    return np.argmax(vector) // 3, np.argmax(vector) % 3


# función principal para jugar contra la IA
def play_game():
    game = tres_en_raya()  # crea una instancia del juego
    model = create_model()  # crea el modelo de la IA

    # ciclo principal del juego
    while True:
        game.display()  # muestra el tablero actual

        # turno del jugador
        try:
            # pide al jugador que ingrese la fila y columna (del 1 al 3)
            row, col = map(int, input("Ingresa tu movimiento (fila, columna del 1 al 3): ").split())
            # verifica que los valores estén entre 1 y 3
            if not (1 <= row <= 3 and 1 <= col <= 3):
                print("Las filas y columnas deben ser del 1 al 3.")
                continue
        except ValueError:
            print("Por favor, ingresa dos números válidos.")
            continue

        # intenta hacer el movimiento del jugador (X)
        if not game.make_move(row, col, 1):
            print("Movimiento inválido. Intenta de nuevo.")
            continue

        # revisa si el juego tiene un ganador
        winner = game.check_winner()
        if winner is not None:
            game.display()  # muestra el tablero final
            if winner == 1:
                print("¡Ganaste!")
            elif winner == -1:
                print("¡La IA ganó!")
            else:
                print("¡Empate!")
            break

        # turno de la IA (O) - movimiento aleatorio
        row, col = random_move(game.board)  # IA elige una casilla al azar
        game.make_move(row + 1, col + 1, -1)  # la IA realiza el movimiento
        print(f"La IA juega en {row + 1}, {col + 1}")  # muestra el movimiento de la IA

        # revisa si la IA ha ganado
        winner = game.check_winner()
        if winner is not None:
            game.display()  # muestra el tablero final
            if winner == 1:
                print("¡Ganaste!")
            elif winner == -1:
                print("¡La IA ganó!")
            else:
                print("¡Empate!")
            break


# iniciar el juego cuando se ejecuta el script
if __name__ == '__main__':
    play_game()  # llama a la función para jugar el juego
```
